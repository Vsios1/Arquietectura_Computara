operadores y asignacion

*opreadores articmeticos 
*opreadores de complejas
*opreadores logicos
*opeadores de relacion

```python
# asignacion

num_favorito = 7
num1 = 7
num2 = 8

print(f'num_favorito = {num_favorito}')

print("operaciones articmeticas")
print(num1,'+',num2, '=',num1+num2) #suma 
print(num1,'-',num2, '=',num1-num2) #resta
print(num1,'*',num2, '=',num1*num2) #multiplicacion
print(num1,'/',num2, '=',num1/num2) #division
print(num1,'//',num2, '=',num1//num2) #division entera
print(num1,'%',num2, '=',num1%num2) #modulo

print("operaciones relacionales")
print(f"{num1} == {num2} = {num1==num2}") #igual
print(f"{num1} != {num2} = {num1!=num2}") #distinto
print(f"{num1} > {num2} = {num1>num2}") #mayor
print(f"{num1} < {num2} = {num1<num2}") #menor
print(f"{num1} >= {num2} = {num1>=num2}") #mayor o igual
print(f"{num1} <= {num2} = {num1<=num2}") #menor o igual-
print(f"{num1} <> {num2} = {num1 != num2}") #distinto

#asignaciones complejas
print("asignaciones complejas")
num1 += num2
print(f"{num1} += {num2} = {num1}")

num2 -= num1
print(f"{num2} -= {num1} = {num2}") 

num1 *= num2
print(f"{num1} *= {num2} = {num1}") 

num1 /= num2
print(f"{num1} /= {num2} = {num1}")     

#operaciones logicas
logico1 = logico2 = False
logico3 = logico4  = True

print(logico1 and logico1)
print(logico1 and logico2)
print(logico1 and logico3)
print(logico1 and logico4)
#and o logico devuelves si almenos hay un tru
print(logico1 or logico1)
print(logico1 or logico2)
print(logico1 or logico3)
print(logico1 or logico4)
#or o logico devuelves si almenos hay un true
```
